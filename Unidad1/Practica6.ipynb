```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el dataset
df = pd.read_csv('/content/autos2.csv')

# Ver las primeras filas
df.head()
```

###Valor mínimo de precios

```python
min_price = df['price'].min()
min_price
```

###Valor máximo de precios

```python
max_price = df['price'].max()
max_price
```

#Moda

```python
mode_price = df['price'].mode()[0]
mode_price
```

###Media

```python
mean_price = df['price'].mean()
mean_price
```

###Mediana

```python
median_price = df['price'].median()
median_price
```

Desviación estándar

```python
std_price = df['price'].std()
std_price
```

###Estadísticos y cuartiles con describe()

```python
df['price'].describe()
```

###Regla de Sturges para determinar bins óptimos

```python
#Usando Formula
n = len(df['price'])
k = 1 + 3.322 * np.log10(n)
k

```

```python
#Para redondear:
bins = int(round(k))
bins
```

###Histograma de precios

```python
plt.figure()
plt.hist(df['price'], bins=bins)
plt.title("Histograma de Precios de Automóviles")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```

###Diagrama de caja:Precio vs Body Style

```python
plt.figure()
sns.boxplot(x='body-style', y='price', data=df)
plt.title("Precio vs Tipo de Vehículo (Body Style)")
plt.xlabel("Tipo de Vehículo")
plt.ylabel("Precio")
plt.xticks(rotation=45)
plt.show()
```

###Diagrama de caja: Precio vs Drive Wheels

```python
plt.figure()
sns.boxplot(x='drive-wheels', y='price', data=df)
plt.title("Precio vs Tipo de Tracción")
plt.xlabel("Tipo de Tracción")
plt.ylabel("Precio")
plt.show()
```

###Diagrama de caja: Precio vs Engine Location

```python
plt.figure()
sns.boxplot(x='engine-location', y='price', data=df)
plt.title("Precio vs Ubicación del Motor")
plt.xlabel("Ubicación del Motor")
plt.ylabel("Precio")
plt.show()
```
