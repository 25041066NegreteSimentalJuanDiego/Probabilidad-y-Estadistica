##Realizar un programa en Python que resuelva el siguiente problema:

##Un fabricante de podadoras para pasto compra motores de 1 hp y 2 ciclos en lotes de 1,000 a un proveedor.
##El fabricante equipa cada una de las podadoras producidas con uno de estos motores.
##Según datos históricos, la probabilidad de que un motor resulte defectuoso es de 0.001.

##En un embarque de 1,000 motores, calcular:

###Cuál es la probabilidad de que ninguno sea defectuoso?
###¿Cuál es la probabilidad de que tres sean defectuosos?
###¿Cuál es la probabilidad de que cuatro sean defectuosos?

##Tipo de distribución

###Esto es binmial, pero también se puede resolver con Poisson para hacerlo más fácil. ¿Por qué binomial? Porque hay un número fijo de pruebas (n = 1000 motores), cada motor puede salir defectuoso o no, y la probabilidad es la misma para todos (p = 0.001). ¿Y Poisson? Como hay muchos motores y la probabilidad de defecto es súper pequeña, se puede usar Poisson con λ = n·p = 1000·0.001 = 1. Básicamente usamos Poison porque simplifica los cálculos y da casi el mismo resultado.


```python
import math
from scipy.stats import binom, poisson

# Parámetros
n = 1000
p = 0.001
lam = n * p  # para Poisson

# a) Ninguno defectuoso (X = 0)
p0_binom = binom.pmf(0, n, p) #Binomial
p0_poisson = poisson.pmf(0, lam) #Poison

# b) Tres defectuosos (X = 3)
p3_binom = binom.pmf(3, n, p)
p3_poisson = poisson.pmf(3, lam)

# c) Cuatro defectuosos (X = 4)
p4_binom = binom.pmf(4, n, p)
p4_poisson = poisson.pmf(4, lam)

# Resultados
print("a) P(X=0)")
print("Binomial:", p0_binom)
print("Poisson :", p0_poisson)

print("\nb) P(X=3)")
print("Binomial:", p3_binom)
print("Poisson :", p3_poisson)

print("\nc) P(X=4)")
print("Binomial:", p4_binom)
print("Poisson :", p4_poisson)
```
