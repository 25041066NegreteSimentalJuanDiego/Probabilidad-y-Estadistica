```python
# Distribución normal de pesos de estudiantes
#Importar
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Mis datos
media = 70
desviacion = 4
total_estudiantes = 500
#----------------------------------------------------------
#Calcular los Limites
x1 = 68
x2 = 72
#-----------------------------------------------------------
# Probabilidad
probabilidad = norm.cdf(x2, media, desviacion) - norm.cdf(x1, media, desviacion)
#--------------------------------------------------------
# Cantidad de estudiantes
cantidad_estudiantes = probabilidad * total_estudiantes
#--------------------------------------------------------
# Resultados
print("Probabilidad de estudiantes que pesan entre 68 kg y 72 kg:")
print(round(probabilidad, 4))

print("\nCantidad aproximada de estudiantes:")
print(round(cantidad_estudiantes))
#----------------------------------------------------
#Grafica
x = np.linspace(55, 85, 1000)
y = norm.pdf(x, media, desviacion)

plt.figure(figsize=(8,5))
plt.plot(x, y, label="Distribución normal")
#----------------------------------------------------
#Esto me sirve para sombrear el area entre 68 y 72
x_fill = np.linspace(x1, x2, 1000)
y_fill = norm.pdf(x_fill, media, desviacion)
plt.fill_between(x_fill, y_fill, alpha=0.5)

plt.axvline(x1, linestyle='--', label="68 kg")
plt.axvline(x2, linestyle='--', label="72 kg")

plt.title("Distribución Normal de los Pesos")
plt.xlabel("Peso (kg)")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.grid()
plt.show()
```
