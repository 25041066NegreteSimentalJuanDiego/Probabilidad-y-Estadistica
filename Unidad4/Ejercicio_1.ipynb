#Probabilidad de que ninguno sufra la enfermedad

```python
from scipy.stats import binom

# Parámetros de la distribución binomial
probabilidad_exito = 0.83
num_pacientes = 16
X = 16

# Calcular la probabilidad de que ninguno de los pacientes sufran la enfermedad
probabilidad_ninguno_enfermo = binom.pmf(X, num_pacientes, probabilidad_exito)

print("La probabilidad de que niguno de los pacientes sufran la enfermedad es:",
      probabilidad_ninguno_enfermo)
```

#Calcular la Probabilidad de que todos sufran la enfermedad:

```python
from scipy.stats import binom

# Parámetros de la distribución binomial
probabilidad_exito = 0.83
num_pacientes = 16
X = 0

# Calcular la probabilidad de que todos los pacientes sufran la enfermedad
probabilidad_todos_enfermos = binom.pmf(X, num_pacientes, probabilidad_exito)

print("La probabilidad de que todos los pacientes sufran la enfermedad es:",
      probabilidad_todos_enfermos)
```
