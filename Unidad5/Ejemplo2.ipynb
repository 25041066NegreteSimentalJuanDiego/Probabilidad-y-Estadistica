```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

#Datos de entrada
publicidad = np.array([2, 3, 5, 4, 11, 5]).reshape(-1, 1)
ganancias = [20, 25, 34, 30, 40, 31]

#Crear modelo y entrenar el modelo de regresion lineal
modelo = LinearRegression()
modelo.fit(publicidad, ganancias)

#Predecir la ganancia para una inversion de 8 millones
inversion_prediccion = np.array([[8]])
ganancia_predicha = modelo.predict(inversion_prediccion)

print("La ganancia esperada para una inversion de 8 en publicidad es: ", ganancia_predicha[0])
print("Valor de la interpretacion", modelo.intercept_)
print("Valor de la pendiente", modelo.coef_)

#Crear el diagrama de dispersion con la linea de regresion
plt.figure(figsize=(6, 5))
plt.scatter(publicidad, ganancias, color='blue', label='Datos')
plt.plot(publicidad, modelo.predict(publicidad), color='red', label='Linea de progresion')
plt.scatter(inversion_prediccion, ganancia_predicha, color='green', marker='x', s=100,
            label='prediccion (8, {0:.2f})'.format(ganancia_predicha[0]))
plt.title('pronostico de ganancias a partir de la inversion en publicidad')
plt.xlabel('inversion en publicidad (millones)')
plt.ylabel('ganancias Anuales (millones) ')
plt.legend()
plt.grid(True)
plt.show()
```
