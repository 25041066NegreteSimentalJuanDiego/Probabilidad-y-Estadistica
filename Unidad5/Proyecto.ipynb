```python
#Imporamos las librrias
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
import numpy as np
```

```python
#Para subir el archivo
from google.colab import files

uploaded = files.upload()
```

```python
#Cargar archivo (datafreame)
df = pd.read_csv("grupo1.csv")

df.head()
```

```python
#Exploración de datos:
#Información general
print(df.info())
```

```python
#Estadistica
print(df.describe())
```

```python
#Histograma de calificaciones
plt.figure(figsize=(8,5))

plt.hist(df["Calificacion"], bins=10)

plt.title("Histograma de Calificaciones")
X = df[["HorasEstudio"]]
Y = df["Calificacion"]

plt.show()
```

```python

#Modelo de regresión lineal
X = df[["HorasEstudio"]]
Y = df["Calificacion"]
#Modelo
modelo = LinearRegression()

modelo.fit(X, Y)

#Modelo de regresión lineal
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

print("Pendiente:", pendiente)
print("Intercepto:", intercepto)
```

```python
#Correlación de Pearson
correlacion, p_valor = pearsonr(df["HorasEstudio"], df["Calificacion"])

print("Coeficiente de correlación:", correlacion)
print("Valor p:", p_valor)
```

```python
#Scatter plot Y la línea de regresión
predicciones = modelo.predict(X)

plt.figure(figsize=(8,5))

plt.scatter(df["HorasEstudio"], df["Calificacion"], label="Datos")

plt.plot(df["HorasEstudio"], predicciones, label="Línea de regresión")

plt.title("Regresión Lineal")
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificación Final")

plt.legend()

plt.show()
```

#Interpretación de Resultados:
##• Analizar los resultados obtenidos en base al diagrama de dispersión la línea de regresión y
##el coeficiente de correlación. Contestar las siguientes preguntas
####1. ¿Explicar cuál es la interpretación del coeficiente de regresión en términos de la relación entre el tiempo de estudio y las calificaciones finales?
#####La pendiente de la recta salió un poco negativa, pero muy pequeña. Eso quiere decir que prácticamente no hay una relación clara entre estudiar más horas y sacar mejores calificaciones en este grupo. En la gráfica se puede ver que la línea está casi plana y los puntos están muy dispersos.
####2. ¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y dirección de la relación entre las variables?
#####El coeficiente de correlación fue de: r=−0.057
#####Ese número está muy cerca de 0, así que la relación entre las horas de estudio y las calificaciones es extremadamente débil. El signo negativo indica una relación negativa, pero realmente es tan pequeña que casi se puede decir que no hay relación lineal entre las dos variables.

####3. ¿El modelo de regresión lineal es estadísticamente significativo?
#####No mucho, porque los datos están muy dispersos y no siguen una tendencia clara. Además, la correlación salió muy cercana a cero, así que el modelo no explica bien las calificaciones usando solamente las horas de estudio.
####4. ¿Qué tan confiable es el modelo de regresión para predecir las calificaciones finales de los estudiantes basado en el tiempo de estudio?
#####El modelo no es muy confiable para hacer predicciones en este grupo. Aunque algunas personas estudiaron más horas, eso no siempre significó mejores calificaciones. Parece que hay otros factores que también afectan mucho el rendimiento académico.
