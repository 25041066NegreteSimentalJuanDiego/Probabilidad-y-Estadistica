```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0,10,20,50])
p = np.array([0.50,0.30,0.15,0.05])

print("Suma:", np.sum(p))

E = np.sum(x*p)
print("Valor esperado:", E)

E2 = np.sum((x**2)*p)
Var = E2 - E**2
std = np.sqrt(Var)

print("Varianza:", Var)
print("Desviación estándar:", std)

F = np.cumsum(p)
print("F(x):", F)

plt.stem(x, p)
plt.title("Distribución de premios")
plt.xlabel("Premio ($)")
plt.ylabel("Probabilidad")
plt.grid()
plt.show()
```

###El valor esperado de 8.5 pesos significa que, en promedio, eso es lo que se ganaría al jugar muchas veces.
###Sin embargo, los resultados varían bastante, ya que algunas veces puedes ganar más, pero muchas otras no ganas nada.
###Por eso, el juego no es muy favorable para el jugador, porque lo más probable es obtener premios bajos o ninguno.
