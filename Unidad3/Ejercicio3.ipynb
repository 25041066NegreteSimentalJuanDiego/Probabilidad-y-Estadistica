```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0,1,2,3,4,5])
p = np.array([0.05,0.10,0.25,0.30,0.20,0.10])

print("Suma:", np.sum(p))

E = np.sum(x*p)
print("Valor esperado:", E)

E2 = np.sum((x**2)*p)
Var = E2 - E**2
std = np.sqrt(Var)

print("Varianza:", Var)
print("Desviación estándar:", std)

F = np.cumsum(p)
print("F(x):", F)

plt.stem(x, p)
plt.title("Ventas por semana")
plt.xlabel("Unidades")
plt.ylabel("Probabilidad")
plt.grid()
plt.show()
```

###En promedio, se venden alrededor de 2.8 productos por semana, lo que muestra un nivel de ventas normal.
###La variación es moderada, así que las ventas no cambian demasiado entre semanas.
###Esto indica que la demanda del producto es bastante estable y no hay cambios muy grandes en las ventas.
