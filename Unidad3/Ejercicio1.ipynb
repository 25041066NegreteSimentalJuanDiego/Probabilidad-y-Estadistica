```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Datos
x = np.array([0,1,2,3,4,5])
p = np.array([0.10,0.25,0.30,0.20,0.10,0.05])

# Verificar suma
print("Suma:", np.sum(p))

# Valor esperado
E = np.sum(x*p)
print("Valor esperado:", E)

# Varianza y desviación estándar
E2 = np.sum((x**2)*p)
Var = E2 - E**2
std = np.sqrt(Var)

print("Varianza:", Var)
print("Desviación estándar:", std)

# Función acumulada
F = np.cumsum(p)
print("F(x):", F)

# Gráfica
plt.stem(x, p)
plt.title("Distribución de probabilidad")
plt.xlabel("X")
plt.ylabel("P(X)")
plt.grid()
plt.show()
```

###“En promedio, cada muestra presenta alrededor de 2 células anormales, lo que sugiere un nivel bajo de alteración celular.”
###“La variabilidad es moderada, lo que indica que aunque la mayoría de las muestras son similares, algunas pueden presentar más células anormales.”
###“La función acumulada muestra que es más probable encontrar pocas células anormales, lo cual puede indicar que la condición estudiada no es severa en la mayoría de los casos.”
